Import danych i czytanie csv

```python
import pandas as pd
breast_cancer = pd.read_csv('breast-cancer-wisconsin.data.csv', names = ["ID", "Clump_Thickness", "Uniformity_of_Cell_Size", "Uniformity_of_Cell_Shape", "Marginal_Adhesion", "Single_Epithelial_Cell_Size", "Bare_Nuclei", "Bland_Chromatin", "Normal_Nucleoli", "Mitoses", "Class"])
```

Odseparowanie klasy od zbioru cech

```python
X = breast_cancer.drop('Class',axis=1)
y = breast_cancer['Class']
```

```python
print (X)
```

```python
print (y)
```

Standaryzacja danych

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
len(X_train)
```

```python
# Fit only to the training data
scaler.fit(X_train)
```

```python
# Now apply the transformations to the data:
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(max_iter=200, verbose=True, hidden_layer_sizes=(100,100),activation='logistic')
```

Opis parametrów wielowarstwowego perceptronu:
https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html

Przykład
mlp = MLPClassifier(max_iter=100, verbose=True, hidden_layer_sizes=(10,10),activation='logistic')
Liczba epok: 100, 2 ukryte warstwy po 10 neuronów, funkcja aktywacji sigmoidalna


```python
mlp.fit(X_train,y_train)
```

```python
predictions = mlp.predict(X_test)
```

```python
predictions
```

```python
y_test
```

WYŚWIETLENIE MACIERZY BŁĘDÓW

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

#prawdziwie pozytywna (ang. true positive, TP)
#prawdziwie negatywna (ang. true negative, TN)
#fałszywie pozytywna (ang. false positive, FP), błąd pierwszego rodzaju
#fałszywie negatywna (ang. false negative, FN), błąd drugiego rodzaju
#czułość (ang. sensitivity, recall) lub odsetek prawdziwie pozytywnych (ang. true positive rate, TPR)
TPR = TP/ P = TP (TP+FN)
#swoistość (ang. specificity, SPC) lub odsetek prawdziwie negatywnych (ang. true negative rate, TNR)
TNR = TN/N = TN /(FP+TN)
#precyzja (ang. precision, lub positive predictive value, PPV – wartość predykcyjna dodatnia)
precyzja = TP/(TP+FP)
#dokładność (ang. accuracy, ACC)
ACC = (TP+TN)/(P+N)
